```python
# Convert to datetime
users['created'] = pd.to_datetime(users.created, unit='ms')
```

```python
users['date'] = users.created.dt.date
users['month'] = users.created.dt.month
```

```python
downloads = pd.DataFrame(users.date.value_counts())
downloads.reset_index(inplace=True)
downloads.rename(columns={"index": "date", "date": "count"},inplace=True)

fig, axs = mpl.pyplot.subplots(ncols=1,nrows=1,figsize=(12,5))

g = sns.lineplot(x="date", y="count",data=downloads)
g.set_title("User Accounts Created over Time",fontsize=20)
```

```python
downloads.sort_values(by='count',ascending=False).head(10)
```

```python
monthly = pd.DataFrame(users.month.value_counts())
monthly.reset_index(inplace=True)
monthly.rename(columns={"index":"Month","month":"Count"}, inplace=True)

fig = sns.barplot(x="Month",y="Count", data=monthly)
fig.set_title("User Accounts Created by Month")
```

```python
data = shares.merge(users[['userId','created','date']],on='userId')
```

```python
data['closed'] = pd.to_datetime(data.closed, unit='ms')
data['ends'] = pd.to_datetime(data.ends, unit='ms')
data['starts'] = pd.to_datetime(data.starts, unit='ms')
data['share_date'] = data.starts.dt.date
data['active_days'] = (data.share_date - data.date).dt.days
```

```python
sns.distplot(data.active_days,hist=True,kde=False)
```

```python
active175 = data[data.active_days < 175]
sns.distplot(active175.active_days,hist=True,kde=False)
```

```python
active50 = data[data.active_days < 50]
sns.distplot(active50.active_days,hist=True,kde=False)
```

```python
active_data = data[['userId','active_days']].groupby(by=data.userId,as_index=False).max()

active175 = active_data[active_data.active_days < 175]
sns.distplot(active175.active_days,hist=True,kde=False)
```

```python
daily_shares = pd.DataFrame(data.share_date.value_counts())
daily_shares.reset_index(inplace=True)
daily_shares.rename(columns={"index":"share_date","share_date":"Count"}, inplace=True)

fig, axs = mpl.pyplot.subplots(ncols=1,nrows=2,figsize=(12,5))

fig1 = sns.lineplot(x="share_date", y="Count", data=daily_shares,ax=axs[0])
fig1.set_title("Shares over Time", fontsize=20)

fig2 = sns.lineplot(x="share_date", y="invitesAccepted", data=data,ax=axs[1])
fig2.set_title("Invites Accepted over Time", fontsize=20)
```
